```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as pt 
import numpy as np
```

```python
iris = pd.read_csv('iris.csv')
```

```python
# (Q) how many datapoints and features?
print(iris.shape)
```

```python
#what are the columns names in of dataset?
print(iris.columns)
```

```python
#how many datapoints are present for each class/species?
iris['Species'].value_counts()
```

```python
#2-D scatter plot
iris.plot(kind='scatter',x='SepalLengthCm',y='SepalWidthCm')
pt.show()
```

```python
#2-D scatter plot  with color-coding for each flower type/class/species
sns.set_style('whitegrid');
sns.FacetGrid(iris,hue='Species',height=4).map(pt.scatter,'SepalLengthCm','SepalWidthCm').add_legend();
pt.show()
```

```python
#pairwise scatter plot : Pair-plot
#can be used with high number of features
#cannot visualize higher dimensional patterns like 3-D or 4-D or more
#only possible to view in 2-D
sns.set_style('whitegrid');
sns.pairplot(iris,hue='Species',height=3)
pt.show()
```

```python
#1-D scatter plot using one feature
#1-D scatter plot using petal-length
iris_setosa = iris.loc[iris['Species'] == 'Iris-setosa'];
#here we are taking iris data frame and all the points that corresponds to setosa and storing it in iris_setosa
iris_virginica = iris.loc[iris['Species'] == 'Iris-verginica'];
#here we are taking iris data frame and all the points that corresponds to virginica and storing it in iris_virginica
iris_versicolor = iris.loc[iris['Species'] == 'Iris-versicolor'];
#here we are taking iris data frame and all the points that corresponds to versicolor and storing it in iris_versicolor
```

```python
#print(iris_setosa['PetalLengthCm'],iris_setosa['PetalWidthCm'])
```

```python
pt.plot(iris_setosa['PetalLengthCm'],np.zeros_like(iris_setosa['PetalLengthCm']))
pt.plot(iris_virginica['PetalLengthCm'],np.zeros_like(iris_virginica['PetalLengthCm']))
pt.plot(iris_versicolor['PetalLengthCm'],np.zeros_like(iris_versicolor['PetalLengthCm']))
```

```python
#plot cdf of petal_length


counts, bin_edges = np.histogram(iris_setosa['PetalLengthCm'],bins=10,density=True)
pdf=counts/(sum(counts))
print(pdf)
print(bin_edges)
```

```python
#compute cdf 
cdf = np.cumsum(pdf)    #cumsum is nothing but cumulative sum
print(cdf)
```

```python
pt.plot(bin_edges[1:],pdf)
pt.plot(bin_edges[1:],cdf)
```
